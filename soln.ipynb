```python
import csv
import collections
import matplotlib.pyplot as plt
with open('amazon_jobs_dataset.csv','r',encoding='UTF-8') as file_obj:
    file_data=csv.DictReader(file_obj,skipinitialspace=True)
    map={}
    for i in file_data:
        date=i['Posting_date'].split()
        if date[2] in map:
            map[date[2]]+=1
        else:
            map[date[2]]=1
    smap=collections.OrderedDict(sorted(map.items()))
    plt.scatter(smap.keys(),smap.values())
    plt.plot(smap.keys(),smap.values())
    plt.draw()
```

```python
import csv
import collections
import matplotlib.pyplot as plt
dict={}
with open('amazon_jobs_dataset.csv','r',encoding='UTF-8') as file_obj:
    file_data=csv.DictReader(file_obj,skipinitialspace=True)
    for i in file_data:
        month=i['Posting_date'].split()[0]
        if month in dict:
            dict[month]+=1
        else:
            dict[month]=1

    plt.bar(dict.keys(),dict.values())
    plt.draw()
    plt.xlabel("Month")
    plt.ylabel("Jobs")
    plt.title('Month vs Job_Openings')
    plt.xticks(rotation = 40)
```

```python
import csv
import collections
import matplotlib.pyplot as plt
dict={}
with open('amazon_jobs_dataset.csv','r',encoding='UTF-8') as file_obj:
    file_data=csv.DictReader(file_obj,skipinitialspace=True)
    map={}
    for i in file_data:
        temp=i['location'].split(',')
        if temp[0]=='IN':
            if temp[2] in map:
                map[temp[2]]+=1
            else:
                map[temp[2]]=1

    for i in map:
        print(i,map[i])
    map=collections.OrderedDict(sorted(map.items()))
    plt.pie(map.values(),labels=map.keys(),autopct='%0.2f')
    plt.title(' Indian cities vs jobs_opening')
    plt.draw()
```
